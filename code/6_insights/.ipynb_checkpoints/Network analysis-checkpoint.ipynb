```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import nxviz as nv
%matplotlib inline
```

```python
CERS = pd.read_csv("/Users/soniawilliams/Desktop/AIS/Archive/CERS_1y.csv",header = 0,delimiter = ',')
```

```python
ports = CERS['POC_LOCODE'].append(CERS['next_port_LOCODE'])
ports = ports.append(CERS['last_port_LOCODE'])
ports.drop_duplicates(keep = 'first', inplace = True)
ports = ports.dropna()
ports = list(ports)
```

```python
voyage1 = CERS.groupby(['last_port_LOCODE','POC_LOCODE'],as_index = False)['voyage_id'].count()
voyage1.dropna()
```

```python
G = nx.DiGraph()
```

```python
for i in ports:
    G.add_node(i)
```

```python
for i, row in voyage1.iterrows():
    G.add_edge(row.last_port_LOCODE, row.POC_LOCODE, weight = row.voyage_id)
```

```python
from nxviz import CircosPlot
```

```python
# Create the CircosPlot object: c
c = CircosPlot(G)
# Draw c to the screen
c.draw()
# Display the plot
plt.show()
```

```python
for n, d in G.nodes(data=True):
    # Calculate the degree of each node: G.node[n]['degree']
    if n[0:2] == 'GB': 
        G.node[n]['grouping'] = 'UK'
    else:
        G.node[n]['grouping'] = 'non-UK'
```

```python
# Create the CircosPlot object: c
d = CircosPlot(G,node_color='grouping',edge_width='weight')
# Draw c to the screen
d.draw()
# Display the plot
plt.show()
```

```python
def get_nodes_and_nbrs(G, nodes_of_interest):
    """
    Returns a subgraph of the graph `G` with only the `nodes_of_interest` and their neighbors.
    """
    nodes_to_draw = []
    
    # Iterate over the nodes of interest
    for n in nodes_of_interest:
    
        # Append the nodes of interest to nodes_to_draw
        nodes_to_draw.append(n)
        
        # Iterate over all the neighbors of node n
        for nbr in G.neighbors(n):
        
            # Append the neighbors of n to nodes_to_draw
            nodes_to_draw.append(nbr)
            
    return G.subgraph(nodes_to_draw)
```

```python
GF = G.subgraph(get_nodes_and_nbrs(G,['GBFXT']))
```

```python
from operator import itemgetter

import matplotlib.pyplot as plt
import networkx as nx

if __name__ == '__main__':
    # Create a BA model graph
    # find node with largest degree
    node_and_degree = GF.degree()
    (largest_hub, degree) = sorted(node_and_degree, key=itemgetter(1))[-1]
    # Create ego graph of main hub
    hub_ego = nx.ego_graph(GF, largest_hub)
    # Draw graph
    pos = nx.spring_layout(hub_ego)
    nx.draw(hub_ego, pos, node_color='b', node_size=50, with_labels=True)
    # Draw ego as large and red
    nx.draw_networkx_nodes(hub_ego, pos, nodelist=[largest_hub], node_size=300, node_color='r')
    plt.show()
```

```python
from __future__ import division

pos = nx.layout.spring_layout(GF)

node_sizes = [3 + 10 * i for i in range(len(GF))]
M = GF.number_of_edges()
edge_colors = range(2, M + 2)
edge_alphas = [(5 + i) / (M + 4) for i in range(M)]

nodes = nx.draw_networkx_nodes(GF, pos, node_size=node_sizes, node_color='blue')
edges = nx.draw_networkx_edges(GF, pos, node_size=node_sizes, arrowstyle='->',
                               arrowsize=10, edge_color=edge_colors,
                               edge_cmap=plt.cm.Blues, width=2)
# set alpha value for each edge
for i in range(M):
    edges[i].set_alpha(edge_alphas[i])

ax = plt.gca()
ax.set_axis_off()
plt.show()
```
